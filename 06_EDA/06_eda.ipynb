```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
titanic
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

1. Scatter Plot

```python
sns.scatterplot( data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

2.Bar Plot

```python
sns.barplot(data=titanic, x='Pclass', y='Age', hue='Sex')
```

3. Box Plot

```python
sns.boxplot(data=titanic, x='Sex', y='Age', hue='Survived')
```

4. Distplot

```python
sns.histplot(data=titanic[titanic['Survived'] == 0], x='Age', kde=True, element='step', stat='density')
sns.histplot(data=titanic[titanic['Survived'] == 1], x='Age', kde=True, element='step', stat='density')
```

5. HeatMap

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

6. ClusterMap 

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived']))
```
