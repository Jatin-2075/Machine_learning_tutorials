```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

df = sns.load_dataset("titanic")
print(df.head())
```

1. Categorical Data

a. countplot

```python
sns.countplot(x='survived', data=df)
plt.show()
```

b. piechart

```python
df['survived'].value_counts().plot(kind = 'pie', autopct='%.2f')
```

```python
plt.hist(df['age'],bins=10)
```

```python
sns.displot(df['age'])
```

```python
sns.boxplot(df['age'])
```
